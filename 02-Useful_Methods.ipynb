```python
#Suppose I have to take the last four digits of the CC number. It is an integer hence we cannot do indexing.
#But, we can convert it to a string and then do the slicing, ex:
str(12345677654321)[-4:]
```

Since, there is no separate function in pandas to do such thing, we can create it as a function and apply separately, using the apply method. 

```python
def last_four(nums):
    return int(str(nums)[-4:])
```

```python
df['last_four']=df['CC Number'].apply(last_four)
```

```python
df
```

```python
df=df.drop('last_four', axis=1)
```

```python
def yelp(price):
    if (price<10):
        return '$'
    elif (price>=10 and price<30):
        return '$$'
    else : 
        return '$$$'
```

```python
df['Yelp']=df['total_bill'].apply(yelp)
# Apply function return single value so our function should also return a single value only. 
```
