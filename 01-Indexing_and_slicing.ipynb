Arrays: Indexing and Slicing

```python
import numpy as np
```

```python
arr= np.arange(0,11)
```

```python
print(arr)
```

```python
arr[5]
```

```python
arr[1:5:2]
```

```python
arr[::]
```

```python
arr[0:5]=100 
# this is not possible to do in a simple list 
```

When we assign a value of the same array to a different object. It points to the same array.
It is visible from this exercise.

```python
arr
```

```python
slice_of_arr=arr[0:5]
```

```python
slice_of_arr[:]=100
```

```python
slice_of_arr
```

```python
arr
```

To avoid changing the values of the original array, we need to make a copy of the original array. 
which can be done using the copy function.

```python
arr_copy=arr.copy()
```

```python
arr_copy[:]=100
```

```python
arr_copy
```

```python
arr
```

Now, we can see that the value of the array didn't get changed. even when we perform any operations on its copy. 

NOW, WE WILL LEARN HOW TO DO THE INDEX AND SELECTION FOR A 2D ARRAY

```python
arr_2d=np.random.randint(0,25,(3,3))
```

```python
arr_2d
```

```python
arr_2d[0]
```

```python
arr_2d[0,2]
# arr_2d[0][2] will do the same thing. 
```

```python
arr_2d[:2,1:]
# We can perform slicing at both rows and columns at once. 
```
