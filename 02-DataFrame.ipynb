To Originally delete from the Data Frame. We can use two options:

1. Inpace command: It is a argument in drop and we can just put df.drop([Column Name],Axis=1,Inplace=True)
2. We can simply assign the new value to object df=df.drop([Column Name], Axis=1)


It is recommended to use 2nd one because it is comaptible with version controls too. 

If we want to delete a row, we can set Axis=0. It will look into rows or if we don't mention anything default of Axis=0.

```python
df=df.drop('Tip percentage',axis=1)
```

```python
df
```

Rows are reffered by Axis=0

Columns are reffered by Axis =1
