SERIES:
A series a data structure in pandas which holds an array of information along with a named index. 
The Named Index differenitates it from Numypy Arrays. 

```python
import numpy as np
```

```python
import pandas as pd
```

Creating a Series:

```python
mydata=[1918,1947,1999,1945]
```

```python
myindex=["WWI","INDIA","KARGIL","WWII"]
```

```python
myser=pd.Series(mydata,myindex)
```

```python
myser
```

IF I wish to create a series from a dictionary. 

```python
mydata2={"Shyam": 22,"Raghav":24,"Diya":20}
```

```python
pd.Series(mydata2)
```

```python
# Imaginary Sales Data for 1st and 2nd Quarters for Global Company
q1 = {'Japan': 80, 'China': 450, 'India': 200, 'USA': 250}
q2 = {'Brazil': 100,'China': 500, 'India': 210,'USA': 260}
```

```python
sales_q1=pd.Series(q1)
```

```python
sales_q2=pd.Series(q2)
```

```python
sales_q1
```

```python
sales_q2
```

We can perform some similar operations on these series like numpy because they are made of numpy as well. 

```python
sales_q1*2
```

Although we cannot do this on a python list or a dictionary.

```python
sales_q1+sales_q2
```

```python
sales=sales_q1.add(sales_q2,0,0)
#This removes the problem of NaN and replaces Nan with a fill value. 
```

```python
sales
```

```python
sales_q1.dtype
```

```python
#Whenever we operate on Series. It gives us result in the format of a float datatype. 
sales.dtype
```

```python
#though we can always change the datatype in the Series function. 
sales.astype(int)
```
